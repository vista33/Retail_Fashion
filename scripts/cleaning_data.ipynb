```python
#1. Import Libraries
import pandas as pd
import numpy as np


```

```python
#2. Load the Dataset
filepath = 'sneakers_streetwear_sales_data.csv'
df = pd.read_csv (filepath)
```

```python
#3. Exploration
print(f"Shape of dataset: {df.shape}")
print ("-"*199)
print(f"\nColumns: {df.columns.tolist()}")
print ("-"*199)
print("\nFirst 5 rows:") 
display(df.head())
print ("-"*199)
print("\nData types:")
display(df.dtypes)
print ("-"*199)
print("\nMissing values:")
display(df.isnull().sum())
```

```python
#4. Handle duplicates
duplicate_exist = df.duplicated().any()
print("Are there duplicate values?")
display (duplicate_exist)

# remove the duplicates in case of any duplicate data
# df = df.drop_duplicates()
```

```python
#5. Standardize column names
df.columns = df.columns.str.replace('[$()]','',regex=True)
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
display(df.head())
```

```python
#6. Convert data types
df['date'] = pd.to_datetime(df['date'], errors = 'coerce')
display(df.dtypes)
display(df.head())
```

```python
#7. Save cleaned data
df.to_csv("cleaned_shoes.csv", index = False)
```
